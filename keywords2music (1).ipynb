```python
import requests
from collections import defaultdict
import random

api_key = 'cfc42c88c2fe92d6e7af878a1090b533'
# tags = ['happy', 'upbeat', 'love', 'pop']
tags = ['rock', 'pop rock', '2017', 'pop']

track_tags = defaultdict(set)
track_details = {}

# Limited by last.fm API itself, each tag can only return up to 1000 songs and the default is to return only 50
limit = 1000

for tag in tags:
    url = f'http://ws.audioscrobbler.com/2.0/?method=tag.gettoptracks&tag={tag}&api_key={api_key}&limit={limit}&format=json'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        if 'tracks' in data and 'track' in data['tracks']:
            tracks = data['tracks']['track']

            # Random: diversify the results of each query
            selected_tracks = random.sample(tracks, min(len(tracks), 500))  # Choose 500 out of 1000

            for track in selected_tracks:
                track_key = (track['name'], track['artist']['name'])
                track_tags[track_key].add(tag)
                track_details[track_key] = track

sorted_tracks = sorted(track_tags.items(), key=lambda x: (-len(x[1]), x[0]))

print("Top tracks across tags with specific matched tags:")
for (track, artist), tags in sorted_tracks[:5]:
    tags_list = ", ".join(sorted(tags))
    print(f"  Track: {track} - Artist: {artist} - Matched Tags: {tags_list}")
```
